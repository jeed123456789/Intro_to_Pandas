male = []
female = []
male_temp = 0 
female_temp = 0
index = 0
numOfYear = 0
while (data['year'][index] == numOfYear + 1987) and (index < 27820):
    if data['sex'][index] == 'male':
        male_temp+=data['suicides_no'][index]
        index+=1
    else:
        female_temp+= data['suicides_no'][index]
        index+=1
    if data['year'][index] != data['year'][index+1]:
        male.append(male_temp)
        female.append(female_temp)

```python
male_tot = []
female_tot = []
for i in range(32):
    male_tot.append(0)
    female_tot.append(0)
for i in range(27820):
    if data['sex'][i] == 'male':
        male_tot[data['year'][i]-1985]+=data['suicides_no'][i]
    else:
        female_tot[data['year'][i]-1985]+=data['suicides_no'][i]
```

```python
male_s = pd.Series(male_tot,name='male')
female_s = pd.Series(female_tot,name='female')
years = []
for i in range(1985,2017):
    years.append(i)
year_s = pd.Series(years,name='year')
```

```python
tot_data = pd.concat([year_s,male_s,female_s],axis=1)
tot_data =tot_data.set_index('year')
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, sharex = True, figsize = (12,4))

# line graph settings
tot_data.male.plot(ax = ax[0], color = 'b', linewidth = 3.0)
tot_data.female.plot(ax = ax[1], color = 'r', linewidth = 3.0)

#labels
fig.suptitle("Annual Worldwide Male and Female Suicides", fontsize = 14, fontweight = "bold") 
ax[0].set_title("Male Suicides", fontsize = 10) 
ax[1].set_title("Female Suicides", fontsize = 10)

for var in ax: 
    var.spines["right"].set_visible(False)
    var.spines["top"].set_visible(False)
    var.set_xlabel("Year")
    var.set_xlim(1985,2014)
ax[0].set_ylim(75000,210000)
ax[1].set_ylim(30000,60000)

ax[0].set_title("Annual Male Suicides") # Lable the y axis (only once since its the same for both)
ax[1].set_title("Annual Female Suicides")
ax[0].legend(["Male Suicides"],frameon=True)
ax[1].legend(["Female Suicides"],frameon=True)

plt.show()
```
